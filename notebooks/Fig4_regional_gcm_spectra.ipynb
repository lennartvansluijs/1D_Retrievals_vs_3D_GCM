# Figure 4 - Regional GCM spectra
This Jupyter notebooks reproduces Figure 4. The output .png and .pdf files are saved in <em>/figures</em>. Regional contributions to the total planet spectrum shown at median observed phase $\phi$ = 0.59.

```python
import pickle
import numpy as np
from mpl_toolkits.basemap import Basemap
from matplotlib import pyplot as plt
import matplotlib.gridspec as gridspec
import scipy.interpolate as spint
from matplotlib import colormaps
```

Load regional spectra.

```python
phase = 0.59
post_processing_run = 'Spec_0_WASP76b_phase_213.75_inc_0.0000.00'
lons_uniq, lats_uniq, line_continuum = pickle.load(open(f'../data/gcm/Regional_Continuum_Flux_{post_processing_run}.pkl', 'rb'))
_, _, line_contrast = pickle.load(open(f'../data/gcm/Regional_Line_Contrast_{post_processing_run}.pkl', 'rb'))
wls_uniq, lats_uniq, lons_uniq, flux_all, solid_all = pickle.load(open(f'../data/gcm/Regional_Spectra_{post_processing_run}.pkl','rb'))

lons, lats = np.meshgrid(lons_uniq,lats_uniq)   
flux_all = flux_all
flux_all /= np.abs(np.max(flux_all))
wls_uniq /= 1e-6
nwl = len(wls_uniq)

def create_spectrum(flux_all, solid_all, wls_uniq, lats_uniq, lons_uniq, initial_lat, final_lat, initial_lon, final_lon):
    total_spectrum = np.zeros(shape=nwl)
    iw = 0
    ilat = 0
    ilon = 0
    big_solid = 0
    count = 0
    
    while iw < len(wls_uniq):
        ilat = 0
        while ilat < len(lats_uniq):
            if initial_lat <= float(lats_uniq[ilat]) <= final_lat:
            #if initial_lat <= float(lats_uniq[ilat]) <= final_lat:
                ilon = 0
                while ilon < len(lons_uniq):
                    #if initial_lon <= float(lons_uniq[ilon]) <= final_lon:
                    if round(float(lons_uniq[0]),0) < initial_lon < round(float(lons_uniq[48]),0):
                        if initial_lon < round(float(lons_uniq[ilon]),0) <= final_lon:
                            total_spectrum[iw] += flux_all[iw, ilat, ilon]
                            if iw == 0:
                                big_solid += solid_all[ilat,ilon]
                                count += 1
                    elif round(float(lons_uniq[0]),0) < initial_lat < round(float(lons_uniq[48]),0): 
                        if initial_lon <= round(float(lons_uniq[ilon]),0) < final_lon:
                            total_spectrum[iw] += flux_all[iw, ilat, ilon]
                            if iw == 0:
                                big_solid += solid_all[ilat,ilon]
                                count += 1
                    else:
                        if initial_lon <= round(float(lons_uniq[ilon]),0) <= final_lon:
                            total_spectrum[iw] += flux_all[iw, ilat, ilon]
                            if iw == 0:
                                big_solid += solid_all[ilat,ilon]
                                count += 1
                    ilon +=1
            ilat += 1
        iw += 1
    total_spectrum *= np.pi * count / (len(lats_uniq) * len(lons_uniq) * big_solid)
    print(count / (len(lats_uniq) * len(lons_uniq)))
    return total_spectrum

def get_index(arr, element):
    """Utility function to get index of element."""
    index = np.argmin(np.abs(arr-element))
    return index

#remove nans before interpolation
line_property = line_continuum
valid_mask = ~np.isnan(line_continuum)  # Mask where values are NOT NaN
valid_points = np.array([lons[valid_mask], lats[valid_mask]]).T
valid_values = line_property[valid_mask]  

# Create query points
query_points = np.array([lons.ravel(), lats.ravel()]).T

#interpolate missing values
interpolated_property = spint.griddata(valid_points, valid_values, query_points, method='cubic')
interpolated_property = interpolated_property.reshape(lons.shape)

# if cubic leaves nans at the edges, use nearest to fill them
if np.isnan(interpolated_property).any():
    interpolated_property = spint.griddata(valid_points, valid_values, query_points, method='nearest').reshape(lons.shape)
    
line_continuum = interpolated_property
line_continuum = line_continuum / np.nanmax(np.abs(line_continuum))

#remove nans before interpolation
line_property = line_contrast
valid_mask = ~np.isnan(line_contrast)  # Mask where values are NOT NaN
valid_points = np.array([lons[valid_mask], lats[valid_mask]]).T
valid_values = line_property[valid_mask]  

# Create query points
query_points = np.array([lons.ravel(), lats.ravel()]).T

#interpolate missing values
interpolated_property = spint.griddata(valid_points, valid_values, query_points, method='cubic')
interpolated_property = interpolated_property.reshape(lons.shape)

# if cubic leaves nans at the edges, use nearest to fill them
if np.isnan(interpolated_property).any():
    interpolated_property = spint.griddata(valid_points, valid_values, query_points, method='nearest').reshape(lons.shape)
    
line_contrast = interpolated_property
line_contrast = line_contrast / np.nanmax(np.abs(line_contrast))
```

Plot figure.

```python
fig = plt.figure(figsize=(7.5,7.5/3))
gs = gridspec.GridSpec(nrows=1, ncols=4, width_ratios=[1.,1.,0.2,0.5], wspace=0.1)
ax1 = fig.add_subplot(gs[0:1,0])
ax2 = fig.add_subplot(gs[0:1,1])
ax3 = fig.add_subplot(gs[0:1,3])

# Line continuum
map = Basemap(projection='ortho',lat_0=0,lon_0=180,resolution='l', ax=ax1)
map.drawparallels(np.arange(-90.,90.,30.))
map.drawmeridians(np.arange(0.,360.,60.))


lons, lats = np.meshgrid(lons_uniq,lats_uniq)
x, y = map(lons, lats)
lonp = [180,180-32,200,160,120]
latp = [0,0,0,-20,0]
colorsp=['tab:green','tab:orange','tab:blue', 'tab:purple', 'tab:red']
xp, yp = map(lonp, latp)

emap = map.contourf(x, y, line_continuum,
                    levels=np.linspace(0, 1, 50),
                    cmap=plt.get_cmap('inferno'))
map.drawmapboundary(fill_color='k')
map.scatter(xp, yp, marker='o', color=colorsp, s=45, alpha=1, zorder=10, lw=0.5, edgecolor='k')

cb = fig.colorbar(emap, location='bottom', aspect=15, pad=0.1, fraction=0.06)
cb.outline.set_linewidth(2)
cb.set_label(label='Local Continuum Flux\n' +r'$\times$ Geometric Factor', fontsize=11)
cb.ax.tick_params(labelsize=11) 
cb.ax.minorticks_on()


# Set 5 tick labels
ticks = np.linspace(0, 1, 5)  # Generate 5 linearly spaced ticks from 1000 to 2000
cb.set_ticks(ticks)  # Set these ticks on the colorbar
cb.set_ticklabels([f"{tick:.2f}" for tick in ticks])

# Line contrast
map = Basemap(projection='ortho',lat_0=0,lon_0=180,resolution='l', ax=ax2)
map.drawparallels(np.arange(-90.,90.,30.))
map.drawmeridians(np.arange(0.,360.,60.))

lons, lats = np.meshgrid(lons_uniq,lats_uniq)
x, y = map(lons, lats)
emap = map.contourf(x, y, line_contrast,
                    levels=np.linspace(-1, 1, 50),
                    cmap=plt.get_cmap('cividis'))
map.drawmapboundary(fill_color='gray')
map.scatter(xp, yp, marker='o', color=colorsp, s=45, alpha=1, zorder=10, lw=0.5, edgecolor='k')

cb = fig.colorbar(emap, location='bottom', aspect=15, pad=0.1, fraction=0.06)
cb.outline.set_linewidth(2)
cb.set_label(label='Local Line Contrast\n' +r'$\times$ Geometric Factor', fontsize=11)
cb.ax.tick_params(labelsize=11) 
cb.ax.minorticks_on()

# Set 5 tick labels
ticks = np.linspace(-1, 1, 5)  # Generate 5 linearly spaced ticks from 1000 to 2000
cb.set_ticks(ticks)  # Set these ticks on the colorbar
cb.set_ticklabels([f"{tick:.1f}" for tick in ticks])

for lop, lap, cp in zip(lonp, latp, colorsp):
    ax3.plot(wls_uniq, flux_all[:, get_index(lats_uniq, lap),  get_index(lons_uniq, lop)], color=cp)


ax3.set_xlabel('Wavelength [$\mu$m]', size=11)
ax3.set_ylabel(r'Local Intensity $\times$ Geometric Factor', size=11)
plt.xlim(2.2954-0.5e-4, 2.2954+0.5e-4)
ax3.set_xticks([2.29535, 2.2954, 2.29545])
ax3.set_xticklabels([2.29535, 2.2954, 2.29545], size=9)

ax3.spines['top'].set_visible(False) 
ax3.spines['right'].set_visible(False)
    
ax3.plot(wls_uniq, np.mean(flux_all, axis=(1,2)), color='k')

for i in range(0, flux_all.shape[1], 1):
    for j in range(0, flux_all.shape[2], 1):
        plt.plot(wls_uniq, flux_all[:,i,j], color='k', lw=0.05, alpha=0.05)
        
plt.plot(wls_uniq, flux_all[:,i,j]*0., color='k', lw=2, alpha=1, label='Average of all lon./lat.')
plt.plot(wls_uniq, flux_all[:,i,j]*0., color='k', lw=0.5, alpha=0.5, label='Set of all lon./lat.')
ax3.legend(bbox_to_anchor=(-0.45, 0.97), loc='lower left', frameon=False)
ax3.set_ylim(0,1)

cmap = colormaps['inferno']
c1 = cmap(0.1)
c2 = cmap(0.9)

ax1.annotate('night', xy=(0.05,0.9), xycoords='axes fraction', fontsize=13, color='w', ha='center', bbox=dict(facecolor=c1, alpha=0.5, edgecolor='k'))
ax1.annotate('day', xy=(0.95,0.9), xycoords='axes fraction', fontsize=13, color='k', ha='center', bbox=dict(facecolor=c2, alpha=0.5, edgecolor=c2))
ax1.annotate('west', xy=(-0.05,0.5), xycoords='axes fraction', fontsize=13, color='k', ha='center', va='center', rotation=90)
ax1.annotate('east', xy=(1.05,0.5), xycoords='axes fraction', fontsize=13, color='k', ha='center', va='center', rotation=-90)
ax1.annotate('north', xy=(0.5,1.05), xycoords='axes fraction', fontsize=13, color='k', ha='center', va='center')
ax1.annotate('south', xy=(0.5,-0.05), xycoords='axes fraction', fontsize=13, color='k', ha='center', va='center')


plt.suptitle(r'Regional Contributions to Total Spectrum ($\phi_{\rm{median}}=$'+f'{phase})', size=11, y=1.175, x=0.5)
plt.savefig(f'../figures/Fig4_regional_gcm_spectra.png', dpi=200, bbox_inches='tight')
plt.savefig(f'../figures/Fig4_regional_gcm_spectra.pdf', bbox_inches='tight')
plt.show()
```
